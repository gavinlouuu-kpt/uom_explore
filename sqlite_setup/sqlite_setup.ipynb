# Deduplicate

```python
def deduplicate_data():
    cursor.execute('''
    DELETE FROM ExperimentData
    WHERE rowid NOT IN (
        SELECT MIN(rowid)
        FROM ExperimentData
        GROUP BY experiment_batch, experiment_id, heater_setting, timestamp
    )
    ''')
```

```python
import os
import sqlite3
import csv
import time

# Path to your database file
db_path = 'D:\\code\\uom_explore\\database\\voc_lab_2.db'

# Function to read CSV and insert data into the database
def insert_data_from_csv(conn, experiment_batch, experiment_id, file_path):
    cursor = conn.cursor()
    print(f"Inserting data from CSV: {file_path}")
    with open(file_path, 'r') as file:
        reader = csv.reader(file)
        first_row = next(reader)
        
        # Check if the first row contains non-numeric values
        try:
            int(first_row[0])
            float(first_row[1])
            float(first_row[2])
            float(first_row[3])
            # First row is valid data, process it
            print(f"Processing first row of data: {first_row}")
            process_row(cursor, experiment_batch, experiment_id, first_row)
        except ValueError:
            # First row is header, skip it and process the rest
            print("First row is a header, skipping")
            pass
        
        for row in reader:
            print(f"Processing row: {row}")
            process_row(cursor, experiment_batch, experiment_id, row)
    
    conn.commit()

def process_row(cursor, experiment_batch, experiment_id, row):
    timestamp, temperature, humidity, pressure = row
    print(f"Inserting row into database: timestamp={timestamp}, temperature={temperature}, humidity={humidity}, pressure={pressure}")
    cursor.execute('''
    INSERT INTO BME_Data (experiment_batch, experiment_id, timestamp, temperature, humidity, pressure)
    VALUES (?, ?, ?, ?, ?, ?)
    ''', (experiment_batch, experiment_id, int(timestamp), float(temperature), float(humidity), float(pressure)))

def process_folders(root_folder):
    conn = sqlite3.connect(db_path)
    try:
        for batch_folder in os.listdir(root_folder):
            batch_folder_path = os.path.join(root_folder, batch_folder)
            if os.path.isdir(batch_folder_path):
                print(f"Processing batch folder: {batch_folder_path}")
                for csv_file in os.listdir(batch_folder_path):
                    if csv_file.endswith('_BME680.csv'):
                        csv_file_path = os.path.join(batch_folder_path, csv_file)
                        experiment_id = os.path.splitext(csv_file)[0]
                        print(f"Found CSV file: {csv_file_path}")
                        insert_data_with_retry(conn, batch_folder, experiment_id, csv_file_path)
    finally:
        conn.close()

def insert_data_with_retry(conn, experiment_batch, experiment_id, file_path, retries=5, delay=1):
    for i in range(retries):
        try:
            insert_data_from_csv(conn, experiment_batch, experiment_id, file_path)
            break
        except sqlite3.OperationalError as e:
            if "database is locked" in str(e):
                print(f"Database is locked, retrying in {delay} seconds...")
                time.sleep(delay)
            else:
                print(f"OperationalError: {e}")
                raise
        delay *= 2

# Root folder containing all batch folders
root_folder = 'D:\\code\\uom_explore\\raw_data\\2024_07_29'

# Process all folders and insert data into the database
print("Starting to process folders...")
process_folders(root_folder)
print("Finished processing folders.")
```
